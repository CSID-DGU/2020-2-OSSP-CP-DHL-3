```python
gather_data = dict()
for gameId in list(games.keys()):
    if np.logical_and(games[gameId]['queueId']==420, 
                      np.logical_and(games[gameId]['mapId']==11, games[gameId]['gameMode']=='CLASSIC')):
        gather_features = dict()
        # win
        wins = dict()
        for j in range(2):
            wins[eval(games[gameId]['teams'])[j]['teamId']] = eval(games[gameId]['teams'])[j]['win']
        # ban
        bans = dict()
        for j in range(2):
            ban_champion_list =[]
            for i in range(5):
                # ë²¤ì„ í•˜ë‚˜ë„ ì•ˆí•œ ê²½ê¸°ë„ ìžˆì„ ìˆ˜ ìžˆë‹¤.
                try:
                    ban_champion_list.append(
                        eval(games[gameId]['teams'])[j]['bans'][i]['championId'])
                except:
                    continue
            bans[eval(games[gameId]['teams'])[j]['teamId']] = ban_champion_list
        # pick & pick&role
        picks = dict()
        pick_details = dict()
        pick_detail_teamId100 = dict()
        pick_detail_teamId200 = dict()
        teamId100 = []
        teamId200 = []
        for i in range(10):
            if eval(games[gameId]['participants'])[i]['teamId'] == 100:
                teamId100.append(eval(games[gameId]['participants'])[i]['championId'])
                pick_detail_teamId100[eval(games[gameId]['participants'])[i]['timeline']['lane'] +'_' + 
                                      eval(games[gameId]['participants'])[i]['timeline']['role']] = \
                                          eval(games[gameId]['participants'])[i]['championId']
            else:
                teamId200.append(eval(games[gameId]['participants'])[i]['championId'])
                pick_detail_teamId200[eval(games[gameId]['participants'])[i]['timeline']['lane'] +'_' + 
                                      eval(games[gameId]['participants'])[i]['timeline']['role']] = \
                                          eval(games[gameId]['participants'])[i]['championId']
        picks[100] = teamId100
        picks[200] = teamId200
        pick_details[100] = pick_detail_teamId100
        pick_details[200] = pick_detail_teamId200
        # gather data
        gather_features['win'] = wins
        gather_features['ban'] = bans
        gather_features['pick'] = picks
        gather_features['pick_role'] = pick_details
        gather_data[gameId] = gather_features
len(gather_data)
```
